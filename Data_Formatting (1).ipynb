```python
import re
def phone_number(num):
    num_str = re.sub(r'\D', '', str(num))
    formatted_number = re.sub(r"(\d{3})(\d{3})(\d{4})", r"\1 \2-\3", num_str)
    return formatted_number
for i in unique_data:
    i["name"]=i["name"].capitalize()
    i["email"]=i["email"].lower()
    i['phone_number']=phone_number(i['phone_number'])
    print(i)
```

```python
def address_method(address):
    temp_data = address.split(',')
    l2 = temp_data[2].split(' ')
    return temp_data[:2] + l2[1:]

d = ['City', 'State', 'Abbreviation', 'Zipcode']
for i in unique_data:
    i.update(dict(zip(d, address_method(i['address']))))

for i in unique_data:
    print(i)
```
